## First get pynams and basic python scientific and plotting packages
For those new to python, I recommend downloading [anaconda](https://www.continuum.io/downloads).  
pynams is written in Python 2.7 and requires the [lmfit](https://pypi.python.org/pypi/lmfit/)
and [uncertainties](https://pythonhosted.org/uncertainties/) packages

```python
from pynams import pynams
import numpy as np
import matplotlib
%matplotlib inline
```

## Each FTIR spectrum needs information about the sample thickness and the file location
I prefer to first specify a sample and keep sample thickness information there

File location is indicated by the folder, which you specify, and a label ('fname', which is also required input for plotting titles and saving data later). In this example the file with the FTIR spectra information is pynamsDemo.CSV

```python
# Make sample with thickness and optional other info
sample1D = pynams.Sample()
sample1D.sample_thick_microns = np.mean([885., 867., 873., 878., 879.])
sample1D.IGSN = 'IEFERPMR1'

# Make spectrum, including sample info and filename
spectrum = pynams.Spectrum()
spectrum.sample = sample1D # specify sample
#spectrum.sample_thick_microns = 8.

### Now get the file with the spectrum. Only .CSV files supported for now.

# Each spectrum requires a label called its fname, which is the file name without the extension. 
# My test data file is called pynamsDemo.CSV, so I set the fname to pynamsDemo. 
# All later files (e.g., peak fits) will use this fname at the start of their filenames.
spectrum.fname = 'pynamsDemo'

## If you have the FTIR file on your computer, you could use code that looks like this
#FTIR_file_location = 'C:\\Users\\Ferriss\\Documents\\FTIR\\'
#pynams.make_filenames(folder=FTIR_file_location, file_ending='.CSV')  

## Here I'm getting my test data from the web and saving it locally to a CSV file
import requests
r = requests.get('https://raw.githubusercontent.com/EFerriss/pynams/master/pynamsDemo.csv')
f = open(''.join((spectrum.fname, '.CSV')),'w')
f.write(r.content)


fig, ax = spectrum.plot_spectrum()
```

## Making baselines
The default is a line between wavenumbers 3200 and 3700 cm-1, but you can also make quadratic baselines and change the wavenumber range. Here are three examples:

```python
fig, ax = spectrum.plot_showbaseline()
```

```python
spectrum.base_low_wn = 3400
baseline = spectrum.make_baseline(line_order=2, shiftline=-0.2, show_plot=True)
```

```python
baseline = spectrum.make_baseline(line_order=2, shiftline=0.3, show_plot=False)
fig, ax = spectrum.plot_showbaseline()
```

## Subtracting the baseline

```python
fig, ax = spectrum.plot_subtractbaseline()
```
