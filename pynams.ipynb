# Pynams: A Python Package for Nominally Anhdydrous Minerals
An example of processing FTIR spectra with pynams from https://github.com/EFerriss/pynams

```python
import pynams
import numpy as np
import matplotlib
%matplotlib inline
```

## Each FTIR spectrum needs information about the sample thickness and the file location
I prefer to first specify a sample and keep sample thickness information there

File location is indicated by the folder, which you specify, and a label ('fname', which is also required input for plotting titles and saving data later). In this example the file with the FTIR spectra information is pynamsDemo.CSV

```python
# Make sample with thickness and optional other info
sample1D = pynams.Sample()
sample1D.sample_thick_microns = np.mean([885., 867., 873., 878., 879.])
sample1D.IGSN = 'IEFERPMR1'

# Make spectrum, including sample info and filename
spectrum = pynams.Spectrum()
spectrum.sample = sample1D # specify sample
FTIR_file_location = 'C:\\Users\\Ferriss\\Documents\\FTIR\\'
spectrum.fname = 'pynamsDemo'
pynams.make_filenames(folder=FTIR_file_location, file_ending='.CSV')  # only .CSV files supported for now

fig, ax = spectrum.plot_spectrum()
```

## You can make linear (default) or quadratic baselines

```python
fig, ax = spectrum.plot_showbaseline()
```

```python
spectrum.base_low_wn = 3400
baseline = spectrum.make_baseline(line_order=2, shiftline=-0.2, show_plot=True)
```

```python
baseline = spectrum.make_baseline(line_order=2, shiftline=0.7, show_plot=False)
fig, ax = spectrum.plot_showbaseline()
```
